```python
import pandas as pd
import matplotlib as mpl
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./Ficheiros/train.csv")
```

```python
df.head()
```

```python
df["model_year"] = df["model_year"].astype()
```

```python
df.info()
```

```python
df['milage'] = df['milage'].str.split(" ", expand = True)[0].str.replace(',', '').astype(int)
```

```python
df.head()
```

```python
df.info()
```

```python
import matplotlib.cm as cm

print(list(cm.datad.keys()))
```

```python
# Seleciona apenas as colunas não numéricas
categorical_columns = df.select_dtypes(include='object').columns

# Mapeia as categorias para números
for column in categorical_columns:
    df[column] = df[column].astype('category').cat.codes

# Calcule a matriz de correlação de Spearman
correlation_matrix = df.corr(method='spearman')

# Visualize a matriz de correlação usando um heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='Greens', fmt=".2f")
plt.title('Matriz de Correlação')
plt.show()
```

```python
fig, axes = plt.subplots(1,2, figsize = (18,6))

sns.histplot(data = df['milage'], ax = axes[0])
sns.scatterplot(data = df, x = 'milage', y = 'price', ax = axes[1])

plt.show()
```

```python
filtered_df = df[df['price'] < 100000]
filtered_df = filtered_df[filtered_df['milage'] < 200000]
```

```python
fig, axes = plt.subplots(1,2, figsize = (18,6))

sns.histplot(data = filtered_df['milage'], ax = axes[0])
sns.regplot(data = filtered_df, x = 'milage', y = 'price', color = 'red',
            scatter_kws = {'color': sns.xkcd_rgb['bluish']},
            line_kws = {'color': sns.xkcd_rgb['dark red'], 'linestyle': '--'}, ax = axes[1])

plt.show()
```

```python
fig, axes = plt.subplots(1,2, figsize = (18,6))

sns.histplot(data = df['model_year'], ax = axes[0])
sns.regplot(data = df[(df['model_year'] >= 1990)], x = 'model_year', y = 'milage', color = 'red',
            scatter_kws = {'color': sns.xkcd_rgb['bluish']},
            line_kws = {'color': sns.xkcd_rgb['dark red'], 'linestyle': '--'}, ax = axes[1])

plt.show()
```

```python
fig, axes = plt.subplots(1,2, figsize = (18,6))

sns.histplot(data = df['model_year'], ax = axes[0])
sns.regplot(data = df[(df['model_year'] >= 1900) & (df['price'] <= 100000)], x = 'model_year', y = 'price', color = 'red',
            scatter_kws = {'color': sns.xkcd_rgb['bluish']},
            line_kws = {'color': sns.xkcd_rgb['dark red'], 'linestyle': '--'}, ax = axes[1])

plt.show()
```

```python
# Seleciona apenas as colunas não numéricas
categorical_columns = df.select_dtypes(include='object').columns
df_engine = df

# Mapeia as categorias para números
for column in categorical_columns:
    df_engine[column] = df_engine[column].astype('category').cat.codes

fig, axes = plt.subplots(1,2, figsize = (18,6))

sns.histplot(data = df_engine['engine'], ax = axes[0])
sns.regplot(data = df_engine[(df['engine'] > -1000)], x = 'engine', y = 'price', color = 'red',
            scatter_kws = {'color': sns.xkcd_rgb['bluish']},
            line_kws = {'color': sns.xkcd_rgb['dark red'], 'linestyle': '--'}, ax = axes[1])

plt.show()
```

```python
fig, axes = plt.subplots(1,2, figsize = (18,6))

sns.barplot(data = df['fuel_type'].value_counts().reset_index(), x = 'fuel_type', y = 'count', color = sns.xkcd_rgb['bluish'], ax = axes[0])
sns.boxplot(data = df[(df['price'] <= 200000)], x = 'fuel_type', y = 'price', color = sns.xkcd_rgb['bluish'], ax = axes[1])

plt.show()
```

```python
fig, axes = plt.subplots(1,1, figsize = (15,5))

sns.histplot(data = df[(df["model_year"]>1990)], x = 'model_year', hue = 'accident', ax = axes)

plt.show()
```

```python
fig, axes = plt.subplots(1,2, figsize = (18,6))

sns.barplot(data = df['clean_title'].value_counts().reset_index(), x = 'clean_title', y = 'count', color = sns.xkcd_rgb['bluish'], ax = axes[0])
sns.boxplot(data = df[(df['price'] <= 200000)], x = 'clean_title', y = 'price', color = sns.xkcd_rgb['bluish'], ax = axes[1])

plt.show()
```
