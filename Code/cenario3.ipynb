# Sem outliers e sem colunas

```python
import pandas as pd
import matplotlib as mpl
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cenario1_engine.csv")
```

```python
df.info()
```

```python
df.head()
```

```python
Q1 = df['model_year'].quantile(0.25)
Q3 = df['model_year'].quantile(0.75)

IQR = Q3 - Q1

limite_inferior = Q1 - 1.5 * IQR
limite_superior = Q3 + 1.5 * IQR

# Filtrar DataFrame para remover outliers
df = df[df['model_year'] <= limite_superior]
df = df[df['model_year'] >= limite_inferior]

# Criar um boxplot com destaque para outliers
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['model_year'], showfliers=True)
plt.title('Boxplot com Outliers')
plt.show()
```

```python
df_final = pd.read_csv("cenario3_engine.csv")
```

```python
# Seleciona apenas as colunas no numricas
categorical_columns = df_final.select_dtypes(include='object').columns

# Mapeia as categorias para nmeros
for column in categorical_columns:
    df_final[column] = df_final[column].astype('category').cat.codes

# Calcule a matriz de correlao de Spearman
correlation_matrix = df_final.corr(method='spearman')

# Visualize a matriz de correlao usando um heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='Blues', fmt=".2f")
plt.title('Matriz de Correlao')
plt.show()
```
