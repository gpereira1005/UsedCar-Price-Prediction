```python
import pandas as pd
import matplotlib as mpl
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./Ficheiros/train.csv")
```

```python
df.head()
```

```python
data = [
    df.shape[0],
    df["clean_title"].isnull().sum(),
    df["clean_title"].nunique(),
    df["clean_title"].duplicated().sum(),
    df["clean_title"].str.isspace().sum(),
]

index=["num_linhas", "nulas", "distintas", "duplicadas", "em branco"]

plt.bar(index, data)

plt.xlabel("Métrica")
plt.ylabel("Valor")

for i in range(len(data)):
    plt.annotate(str(data[i]), (index[i], data[i]))

plt.show()
```

```python
df.head()
```

```python
clean_title = df['clean_title'].value_counts()[:2]

# Criar um gráfico de barras
plt.figure(figsize=(3, 5))
plt.bar(clean_title.index, clean_title)

for i in range(len(clean_title)):
    plt.annotate(str(clean_title[i]), (clean_title.index[i], clean_title[i]))

plt.show()
```
