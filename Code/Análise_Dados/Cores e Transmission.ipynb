```python
import pandas as pd
import matplotlib as mpl
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../Ficheiros/train.csv")
```

```python
data = [
    df.shape[0],
    df["ext_col"].isnull().sum(),
    df["ext_col"].nunique(),
    df["ext_col"].duplicated().sum(),
    df["ext_col"].str.isspace().sum(),
]

index=["num_linhas", "nulas", "distintas", "duplicadas", "em branco"]

plt.bar(index, data)

plt.xlabel("Métrica")
plt.ylabel("Valor")

for i in range(len(data)):
    plt.annotate(str(data[i]), (index[i], data[i]))

plt.show()
```

```python
# Obter os 5 valores mais frequentes

ext_col = df['ext_col'].value_counts()[:5]

# Criar um gráfico de barras
plt.figure(figsize=(12, 4))
plt.bar(ext_col.index, ext_col)

for i in range(len(ext_col)):
    plt.annotate(str(ext_col[i]), (ext_col.index[i], ext_col[i]))

plt.show()
```

```python
df.loc[df['transmission'].str.contains('CVT'), 'tsm'] = 'CVT'
df.loc[df['transmission'].str.contains('automatic|Automatic|A/T|AT|Auto'), 'tsm'] = 'Automatic'
df.loc[df['transmission'].str.contains('M/T|Manual|Mt'), 'tsm'] = 'Manual'
df.loc[df['transmission'].str.contains('Dual Shift Mode'), 'tsm'] = 'DCT'
df.loc[df['tsm'].str.contains('nan'), 'tsm'] = 'Outras'
df['tsm'].fillna('Outras')
```
